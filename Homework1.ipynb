```python
import matplotlib.pyplot as plt
import pandas as pd

#retrieves and reads the data frame from github
url = 'https://vincentarelbundock.github.io/Rdatasets/csv/ggplot2/diamonds.csv'
df = pd.read_csv(url, index_col= 0)

#Prints the first 7 rows
print(df.head(7))

```

```python
#Prints the last 7 rows
print(df.tail(7))
```

```python
#Prints the statistics for the data frame
print(df.describe())
```

```python
#Prints the colums cut, color, and clarity and get statistics from it.
print(df[['cut', 'color', 'clarity']].describe())
```

```python
#Print the unique values in each column
print("Unique values in 'cut':", df['cut'].unique())
print("Unique values in 'color':", df['color'].unique())
print("Unique values in 'clarity':", df['clarity'].unique())
```

```python
#Makes histogram of carat
plt.hist(df['carat'])

plt.title('Histogram of carat')

plt.show()
```

```python
#Makes histogram of price
plt.hist(df['price'])

plt.title('Histogram of price')

plt.show()
```

```python
#Makes histogram of table
plt.hist(df['table'])

plt.title('Histogram of table')

plt.show()
```
